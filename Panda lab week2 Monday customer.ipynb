```python
import pandas as pd
import numpy as np
import statistics as stat
```

```python
file1_df= pd.read_csv("file1.csv")
file1_df
```

```python
file2_df= pd.read_csv("file2.csv")
file2_df
```

```python
file3_df= pd.read_csv("file3.csv")
file3_df
```

```python
#df.columns = df.columns.str.replace( " ", "_") replacing 
```

```python
file1_df = file1_df.rename(columns={'ST':'State','GENDER':'Gender'})
file1_df
```

```python
file2_df = file2_df.rename(columns={'ST':'State','GENDER':'Gender'})
file2_df
```

```python
concat_df = pd.concat([file1_df,file2_df, file3_df], axis=0)
concat_df
```

```python
#2. Standardizing header names

concat_df.columns = concat_df.columns.str.replace(" ", "_") #replacing
concat_df
```

```python
#3. Removing duplicates

concat_df.drop_duplicates(inplace=True)
concat_df
```

```python
# 4. Deleting and rearranging columns – delete the column customer as it is only a unique identifier for each row of data

concat_df.drop(columns=['Customer'], inplace=True)
concat_df
```

```python
#5. Working with data types – Check the data types of all the columns and fix the incorrect ones (for ex. customer lifetime value and number of open complaints ). Hint: remove the percentage from the customer lifetime value and truncate it to an integer value.


concat_df.dtypes
```

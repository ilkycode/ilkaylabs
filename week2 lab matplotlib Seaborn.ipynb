```python
import pandas as pd
import numpy as np
import statistics as stat
import random as rand 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
customer_df= pd.read_csv("Data_Marketing_Customer_Analysis_Round3.csv")
customer_df
```

```python
#- Get the numeric data into dataframe called `numerical` and categorical columns in a dataframe called `categoricals`.
#(You can use np.number and np.object to select the numerical data types and categorical data types respectively)

numericals = customer_df.select_dtypes(include=[np.number])
categoricals = customer_df.select_dtypes(exclude=[np.number])

#numeric_data.shape[1]
#categorical_data.shape[1] 
numericals
```

```python
categoricals
```

```python
# - Use seaborn library to construct distribution plots for the numerical variables
#check normality with histogram in seaborn and matplotlib, you can use the hist func
```

```python
sns.distplot(numericals)
```

```python
# - Use Matplotlib to construct histograms
customer_df.hist(figsize=(10,15))

#none of them looks like normal distribution below.
```

```python
rel_claim_amount = customer_df.groupby(['income','education'])['total_claim_amount'].count().reset_index()
rel_claim_amount
#here I am looking for how income and open number of complaints affects the total claim amount.
```

```python
sns.set_palette('RdYlGn') 
palette_education = {"bachelor": "C5", "high school or below": "C0","college": "C4", "doctor": "C3","master": "C2"}
```

```python
plt.figure(figsize=(9,8))
plt.subplot(2,2,1)
plt.title('Total claim amount by education', size='15', fontweight='bold')

sns.barplot(x = rel_claim_amount['income'],y = rel_claim_amount['total_claim_amount'], hue = rel_claim_amount['education'], palette=palette_education)
```

```python
rel_claim_amount = customer_df.groupby(['income','region'])['total_claim_amount'].count().reset_index()
rel_claim_amount
```

```python
sns.set_palette('RdYlGn') 
palette_region = {"central": "C5", "west region": "C0","east": "C4","north west": "C3"}
```

```python
plt.figure(figsize=(9,8))
plt.subplot(2,2,1)
plt.title('Total claim amount by region', size='10', fontweight='bold')

sns.barplot(x = rel_claim_amount['income'],y = rel_claim_amount['total_claim_amount'], hue = rel_claim_amount['region'], palette=palette_region)
```
